```python
### matching the action with bounding box
time1 = csv.DictReader(file('video/video_living_room_check.csv', 'rb'))
time2 = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
time3 = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
action = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
t1 = [row['t'] for row in time1]
t2 = [row['start'] for row in time2]
t3 = [row['end'] for row in time3]
a = [row['name'] for row in action]
time_sequence1 = []
time_sequence2 = []
time_sequence3 = []
action_sequence = []
for i, val in enumerate(t1):
    time_sequence1.append(int(val))
for i, val in enumerate(t2):
    time_sequence2.append(float(val))
for i, val in enumerate(t3):
    time_sequence3.append(float(val))
for i, val in enumerate(a):
    action_sequence.append((val))
#print(action_sequence)

csvfile_input = file('video/video_living_room_check.csv', 'rb')
csvfile_output = file('video/video_living_room_final_index.csv', 'wb')
reader = csv.reader(csvfile_input)
writer = csv.writer(csvfile_output)

all = []
row = next(reader)
row.append('action')
all.append(row)
index = 0
for row in reader:
    for i, val in enumerate(time_sequence2):
         if(float(row[0])>=float(val) and float(row[0])<=float(time_sequence3[i])):                        
                index = i                   
                break
         else:
                continue
    action = action_sequence[index]
    row.append(action)
    all.append(row)
writer.writerows(all)
csvfile_input.close()
csvfile_output.close()
```

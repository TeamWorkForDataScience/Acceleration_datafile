```python
### matching the action with bounding box
time1 = csv.DictReader(file('video/video_living_room_check.csv', 'rb'))
time2 = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
time3 = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
action = csv.DictReader(file('video/annotations_0_resample.csv', 'rb'))
t1 = [row['t'] for row in time1]
t2 = [row['start'] for row in time2]
t3 = [row['end'] for row in time3]
a = [row['name'] for row in action]
time_sequence1 = []
time_sequence2 = []
time_sequence3 = []
action_sequence = []
for i, val in enumerate(t1):
    time_sequence1.append(int(val))
for i, val in enumerate(t2):
    time_sequence2.append(float(val))
for i, val in enumerate(t3):
    time_sequence3.append(float(val))
for i, val in enumerate(a):
    action_sequence.append((val))
#print(action_sequence)

csvfile_input = file('video/video_living_room_check.csv', 'rb')
csvfile_output = file('video/video_living_room_final_index.csv', 'wb')
reader = csv.reader(csvfile_input)
writer = csv.writer(csvfile_output)

all = []
row = next(reader)
row.append('action')
all.append(row)
index = 0
for row in reader:
    for i, val in enumerate(time_sequence2):
         if(float(row[0])>=float(val) and float(row[0])<=float(time_sequence3[i])):                        
                index = i                   
                break
         else:
                continue
    action = action_sequence[index]
    row.append(action)
    all.append(row)
writer.writerows(all)
csvfile_input.close()
csvfile_output.close()
```

```python
elements = df.copy()
#elements = elements[elements["atomic number"] <= 82]
#elements = elements[~pd.isnull(elements["melting point"])]
#mass = [float(x.strip("[]")) for x in elements["atomic mass"]]
#elements["atomic mass"] = mass
palette = ["#053061"]
original_length = elements["2D_length"]
original_width = elements["2D_wide"]
length = []
width = []
area = []
time = []
height = []
a = []
for i, val in enumerate(original_length):
    val = val
    length.append(val)

for i, val in enumerate(original_width):
    val = val
    width.append(val)
    
for i, val in enumerate(elements["2D_area"]):
    val = val
    area.append(val)
for i, val in enumerate(elements["t"]):
    val = val
    time.append(val)
for i, val in enumerate(elements["3D_height"]):
    val = val
    height.append(val)
for i, val in enumerate(elements["action"]):
    val = val
    a.append(val)




#melting_points = elements["melting point"]
#low = min(melting_points)
#high= max(melting_points)

#melting_point_inds = [int(10*(x-low)/(high-low)) for x in melting_points] #gives items in colors a value from 0-10
#elements['melting_colors'] = [palette[i] for i in melting_point_inds]

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"
TITLE = "2D_bounding_box"

p = figure(tools=TOOLS, toolbar_location="above", logo="grey", plot_width=1200, title=TITLE)
p.background_fill_color= "#dddddd"
p.xaxis.axis_label="time"
p.yaxis.axis_label="2D_area"
p.grid.grid_line_color="white"

partTimeData = time[0:100]
partAreaData = area[0:100]
partActionData = a[0:100]

source = ColumnDataSource(data=dict(t = partTimeData, h = partAreaData, action = partActionData))
for n in range(len(partTimeData)):
    p.rect(x=time[n], y=area[n], width=width[n]/25, height=length[n]/25, color="#CAB2D6",
           line_color="black",  angle=pi/2, line_width = 0.5, height_units="screen")
    
    
    
labels = LabelSet(x='t', y='h', text='action', y_offset=8,
                  text_font_size="8pt", text_color="#555555",
                  source=source, text_align='center')
p.add_layout(labels)

    
#output_file("elements.html", title="elements.py example")


show(p)
```
